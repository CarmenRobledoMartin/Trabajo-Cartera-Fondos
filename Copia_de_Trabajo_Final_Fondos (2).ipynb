Este código realiza una consulta en Google BigQuery para obtener datos de valor liquidativo (NAV) de activos financieros identificados por códigos ISIN

```python
# Proyecto en el que se ejecuta.
project = 'ironia-data'

#Creacion  del cliente

client = bigquery.Client(project=project, credentials=credentials)
sql_str = "SELECT * FROM( SELECT date,isin,nav FROM ironia-data.ironia.navs WHERE date BETWEEN '2020-01-01' AND CURRENT_DATE()) PIVOT (MAX(nav) FOR isin IN (" + ListaISIN + ")) ORDER BY Date ASC"
```

```python
ListaISIN = ['IE00BNQ4RL64','LU0570466788','LU0945454980'] # Crear una lista de los ISIN selccionados
pesos = np.ones(len(ListaISIN)) / len(ListaISIN)  # Hacer pesos equiponderados
pesos

# Lista para almacenar rendimientos de la cartera
rentabilidades_fondo = []
pesos_fondo = []
volatilidad_fondo = []
rng = np.random.default_rng(seed=42)
for idx in range(1):
    # Filtrar los datos de rendimientos logarítmicos para los activos seleccionados
    fondo_ret = datos_fondos[ListaISIN]
    pesos=np.ones(len(ListaISIN)) / len(ListaISIN)

    # Calcular el rendimiento de la cartera
    rentabilidad_fondo = fondo_ret @ pesos    #Se calcula la rentabilidad de la cartera
    rentabilidades_fondo.append(rentabilidad_fondo)  #Se almacena los rendimientos de la cartera en la lista creada anteriormente.
    pesos_fondo.append(pesos)

    #Calculo volatilidad
    volat_fondo = rentabilidad_fondo.std()
    volatilidad_fondo.append(volat_fondo)

# Resultado final: rendimiento de la cartera
for i, ret in enumerate(rentabilidades_fondo):
    print(f"\nRendimientos de la cartera")
    print(ret)

for i, vol in enumerate(volatilidad_fondo):
    print(f"\nVolatilidad de la cartera")
    print(vol)

for i, w in enumerate(pesos):
  print(f'Cartera{i + 1}: {dict(zip(ListaISIN, pesos))}')
```

```python
cartera_df = pd.DataFrame(rentabilidades_fondo).T

cartera_df.columns = [f'Cartera_{i+1}' for i in range(1)]


def calcular_cvar(rendimientos, alpha=0.05):
    """
    Calcula el CVaR (Expected Shortfall) para una serie de rendimientos.
    - rendimientos: Series de rendimientos de la cartera.
    - alpha: Nivel de confianza, típicamente 0.05 (para el 5% de VaR)
    """
    # Calcular el VaR: percentil inferior de los rendimientos (VaR en el nivel de confianza alpha)
    var = np.percentile(rendimientos, 100 * alpha)
    # Calcular el CVaR: promedio de las pérdidas que son menores que el VaR
    cvar = rendimientos[rendimientos <= var].mean()

    return cvar

# Calcular el CVaR para la cartera
cvars = {}
for i in range(1):
    cvars[f'Cartera_{i+1}'] = calcular_cvar(cartera_df[f'Cartera_{i+1}'])
cvar_lista = []

# Mostrar el CVaR calculado
print("\nCVaR de la cartera:")
for cartera, cvar in cvars.items():
    print(f"{cartera}: {cvar}")
    cvar_lista.append(cvar)

#Mostrar la cartera con el CVaR más pequeño
minima_cvar = -1000

# Mostrar el CVaR calculado
for cartera, cvar in cvars.items():
  if cvar > minima_cvar:
    minima_cvar = cvar
    cartera_con_minimo_cvar = cartera
    print(f'\nLa cartera tiene un CVar de {cvar_lista}')
```
